```python
#ploting in pandas
graph = pd.Series(np.random.normal(size=30))
print(graph)
graph.plot()
```

```python
graph.cumsum().plot(grid=True)
```

```python
#for dataFrame
variables = pd.DataFrame({'normal':np.random.normal(size=10),
                         'gamma': np.random.gamma(1,size=10),
                         'poisson':np.random.poisson(size=10)})
print(variables)

variables.plot()
```

```python
variables.cumsum(0).plot(subplots = True)
```

```python
variables.plot(secondary_y='normal')
```

```python
fig,axes = plt.subplots(nrows = 3,ncols=1,figsize = (3,10))
for i , var in enumerate(['normal','gamma','poisson']):
    variables[var].cumsum(0).plot(ax = axes[i],title = var)
axes[0].set_ylabel('y-axis')
```

```python
#to save figure
fig.savefig('figure.png')
```

```python
fig = plt.figure()
ax = plt.axes()
s  = np.linspace(0,20,100)
print(s)
ax.plot(s,np.sin(s));
```

```python
plt.plot(s,np.sin(s))
```

```python
#line color 
plt.plot(s,np.sin(s-0),color = 'red')
plt.plot(s,np.sin(s-1),color = 'y')
plt.plot(s,np.sin(s-2),color = '0.55')
```

```python
#line styles
plt.plot(s,np.sin(s-0),Linestyle = '-')
plt.plot(s,np.sin(s-1),Linestyle = ':')
plt.plot(s,np.sin(s-2),Linestyle = '-.')
```

```python
plt.plot(s,np.sin(s-0))
plt.xlim(-1,15)
plt.ylim(-2,2)
```

```python
#labeling the points
plt.plot(s,np.cos(s))
plt.title('cosine curve')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.style.use('seaborn-whitegrid')
```

```python
rng = np.random.RandomState(10)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 100*rng.rand(100)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap='Reds')
plt.colorbar();
```

```python
x = np.linspace(0,10,100)
dy = 0.3
y = np.sin(x)+dy*np.random.randn(100)
plt.errorbar(x,y,yerr=dy,fmt = 'r');
```

```python
mean = [5,2]
cov = [[3,1],[2,0.5]]
x,y = np.random.multivariate_normal(mean,cov,100000).T
```

```python
plt.hist2d(x,y,bins = 20,cmap = 'Accent_r')
cb = plt.colorbar()
```
